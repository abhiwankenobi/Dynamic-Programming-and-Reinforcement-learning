```python
import numpy as np
import math
```

```python
class part_one:
    def qlearning():
        states = np.arange(10) #we can have 5 or 10 states
        alpha = 0.5 #learning rate
        epsilon = 1 
        gamma = 0.9 #discount rate
        episodes = 1000 #epsilon iterations
        iterations = 1000 #no.of times we run the program
        n=episodes+1 
        decay_start = 1 #the point where our epsilon starts decaying
        decay_end = math.floor(episodes/1.5) #the point where our decay ends
        epsilon_decay = epsilon/(decay_end - decay_start) #amount of decay on epsilon value
        long_run_avg = {' Episode no ': [], ' Average ': []} #dictionary for all the avg values
        q_table = np.zeros((10,2))

        for i in range(n):
            rewards = []
            init_state = 0
            for j in range(iterations):
                if np.random.random() > epsilon: #we check the condition for epsilon greedy
                    action = np.argmax(q_table[init_state,:])
                else:
                    action = np.random.randint(2)
                if action == 1: 
                    next_state = 0
                    if init_state == 0:
                        r=0.2 #reward at first state (b)
                    else: 
                        r=0 #reward at intermediate states
                else:
                    if init_state == states[-1]:
                        next_state = init_state
                        r=1 #reward at last state (a)
                    else :
                        next_state = init_state+1
                        r=0 #reward at intermediate states
                rewards.append(r)
                q_dash = np.max(q_table[next_state,:])
                q = q_table[init_state,action]
                q_update = (1 - alpha) * q + alpha * (r + gamma * q_dash) #qlearning formula
                q_table[init_state,action] = q_update

                if i % 10 == 0 and j == iterations - 1:
                    print("Episode ", i, ":\n", q_table)

                init_state = next_state

            r_avg = sum(rewards)/iterations
            long_run_avg[' Episode no '].append(i)
            long_run_avg[' Average '].append(r_avg)
#             long_run_avg[' Maximum '].append(max(rewards))
#             long_run_avg[' Minimum '].append(min(rewards))
            if not i%5:
                print(f'Episode: {i:>4d}, long run average reward: {r_avg:>4.1f}, current epsilon: {epsilon:>1.2f}')
            if decay_end >= i >= decay_start:
                epsilon -= epsilon_decay
    
```

```python
if __name__ == "__main__":
    qtable = part_one
    qtable.qlearning()
```
